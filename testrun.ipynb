```python
commonDict =  {}
for ing in commonIng: commonDict[ing] = {}
```

```python
def Adder(dic,keyList):
    for key in keyList:
        if key in dic: dic[key] += 1
        else: dic[key] = 1
```

```python
for lis in dataset["ingredients"]:
    clean_list = [item for item in lis if item not in commonIng]
    for common in commonIng:
        if common in lis:
            Adder(commonDict[common],clean_list)
    
```

```python
maxDict = copy.deepcopy(commonDict)
for common in commonDict:
    maxs = []
    for i in range(20):
        max_ing = max(maxDict[common], key=maxDict[common].get)
        maxDict[common].pop(max_ing)
        maxs.append(max_ing)
    maxDict[common] = maxs
```

```python
for common in maxDict:
    print(common+":",','.join(maxDict[common]))
```

```python
dfs = []
for common in maxDict:
    df = pd.DataFrame(dataset['cuisine'])
    df[common] = pd.Series(dtype="bool")
    #df = df.reindex(columns=df.columns.tolist()+maxDict[common])
    for ing in maxDict[common]: df[ing] = pd.Series(dtype="bool")
    for i in range(df.shape[0]):
        ings = dataset.iat[i,2]
        if common in ings: df.iat[i,1] = True
        for ing in range(len(maxDict[common])):
            if maxDict[common][ing] in ings: df.iat[i,ing+2] = True
            else: df.iat[i,ing+2] = False
    dfs.append(df)
```

```python
trueDict = {}
falseDict = {}
for i in range(dfs[0].shape[0]):
    if dfs[0].iat[i,1]==True:
        val = sum(1 for x in range(20) if dfs[0].iat[i,x+2]==True)
        if val in trueDict: trueDict[val] += 1
        else: trueDict[val] = 1
    else:
        val = sum(1 for x in range(20) if dfs[0].iat[i,x+2]==True)
        if val in falseDict: falseDict[val] += 1
        else: falseDict[val] = 1
```
